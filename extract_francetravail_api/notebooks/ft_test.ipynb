```python
import os
import requests
from pprint import pprint
from dotenv import load_dotenv

load_dotenv()
CLIENT_ID = os.getenv('FT_CLIENT_ID')
CLIENT_SECRET = os.getenv('FT_CLIENT_SECRET')
SCOPE = os.getenv('SCOPE_API_FT_EMPLOI')
TOKEN_URL = os.getenv('TOKEN_URL')
BASE_URL = os.getenv('BASE_URL')
API_CODE = os.getenv('API_CODE')
API_VERSION = os.getenv('API_VERSION')
RESOURCE = os.getenv('RESOURCE')
REQUEST_URL = f'{BASE_URL}/{API_CODE}/{API_VERSION}/{RESOURCE}'
print('Endpoint:', REQUEST_URL)
assert CLIENT_ID and CLIENT_SECRET and TOKEN_URL, 'Variables FT manquantes'
```

```python
def get_access_token():
    data = {
        'grant_type': 'client_credentials',
        'client_id': CLIENT_ID,
        'client_secret': CLIENT_SECRET,
        'scope': SCOPE,
    }
    resp = requests.post(TOKEN_URL, data=data, timeout=10)
    resp.raise_for_status()
    payload = resp.json()
    print('Token OK, expires_in:', payload.get('expires_in'))
    return payload['access_token']

TOKEN = get_access_token()

```
