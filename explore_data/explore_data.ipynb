```python
def get_keys(data):
    result = set()
    if isinstance(data,dict):
        for k, v in data.items():
            result.add(k)
            result.update(get_keys(v))
        return result
    elif isinstance(data,list):
        for i in data:
            result.update(get_keys(i))
        return result
    else:
        return set()

def flatten_json(y, delimiter = '.'):
    out = {}
    
    def flatten(x, name=''):
        if isinstance(x,dict):
            for k,v in x.items():
                flatten(v, name + k + delimiter)
        elif isinstance(x,list):
            for i, v in enumerate(x):
                flatten(v, name + f"_{str(i)}" + delimiter)
        else:
            # on enlève le dernier delimiter
            out[name[:-len(delimiter)]] = x 

    flatten(y)
    return out

def norm_json(data):
    all_keys = set()
    
    for e in data:
        if isinstance(e,dict):
            all_keys.update(e.keys())
        else:
            print("Not a dictionnary.")
            return None
    
    for e in data:
        for k in all_keys:
            if k not in e.keys():
                e[k] = None

    return data

```
